#how should we replace the nan or missing values

```python
 data=df.copy()
for features in feature_with_nan:
   
    #1=inplace of nan value , 0= inplace of not nan value, in all col that have nan value
    data[features]=np.where(data[features].isnull(),1,0)
    #calculate the mean sales price(our output),for nan values and non nan values
    data.groupby(features)['SalePrice'].median().plot.bar()
    plt.title(features)
    plt.show()
    

    
```

#Two types of nume variables-
Discrte and Continuous variable(if less unique values in that feature then it is called as discrete feature)

```python
feature=[feature for feature in df.columns]
discrte_feature=[]
for i in feature:
    if df[i].dtype!='object':
        if len(df[i].unique())<=25:
            discrte_feature.append(i)
print(discrte_feature)       
    
```

```python
len(discrte_feature)
```

```python
df[discrte_feature].head()
```

#find relationship bet discrete var and sales price

```python
data=df.copy()
for fea in discrte_feature:
    data.groupby(fea)['SalePrice'].median().plot.bar()
    plt.xlabel(fea)
    plt.ylabel('SalePrice')
    plt.show()
```

```python
con_fea=[]
for i in feature:
    if df[i].dtype!='object':
        if i not in discrte_feature and i not in date_time:
            con_fea.append(i)
print(con_fea)        
```

```python
con_fea.remove('Id')
len(con_fea)
```

#analyse the rel bet con feature and saleprice by using histogram

```python
data=df.copy()
for i in con_fea:
    data[i].plot.hist()
    plt.xlabel(i)
    plt.ylabel('count')
    plt.show()
```
